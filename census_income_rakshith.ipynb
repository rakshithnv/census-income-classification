```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('census-income.csv')
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
df.head()
```

```python
df.columns
```

```python
df['native-country']=df['native-country'].str.replace(' ','')
```

```python
df['native-country'].unique()
```

```python
df.info()
```

```python
def remove_spaces(dataframe):
  for i in dataframe.columns:
    if dataframe[i].dtype == 'object':
      dataframe[i]=dataframe[i].str.replace(' ','')
```

```python
remove_spaces(df)
```

```python
df.head()
```

```python
df['marital-status'].unique()
```

```python
df.isnull().sum()
```

```python
df=df.replace('?',np.nan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.columns
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
df.info()
```

```python
df.rename(columns={'':'annual-income'},inplace=True)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
```

```python
y=df['annual-income']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

**Logistic regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
#testing
lr_pred=lr.predict(x_test)
```

```python
lr_pred
```

```python
y_test
```

```python
from sklearn.metrics import *
```

```python
lr_accuracy=accuracy_score(y_test,lr_pred)
```

```python
lr_accuracy
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model=DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
dt_pred=dt_model.predict(x_test)
```

```python
dt_pred
```

```python
dt_acc=accuracy_score(y_test,dt_pred)
```

```python
dt_acc
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier()
```

```python
rf_model.fit(x_train,y_train)
```

```python
rf_pred=rf_model.predict(x_test)
```

```python
rf_pred
```

```python
rf_acc=accuracy_score(y_test,rf_pred)
```

```python
rf_acc
```

```python
print("Accuracy of LogisticRegression",lr_accuracy)
print("Accuracy of Decision Tree",dt_acc)
print("Accuracy of Random Forest",rf_acc)
```

```python
trees=[i for i in range(50,800,50)]
```

```python
trees
```

```python
# Initialize variables to keep track of the best model and its accuracy
best_n = None
best_accuracy = 0
best_rf_model = None
best_rf_pred = None
for n in trees:
  rf_model=RandomForestClassifier(n_estimators=n, random_state=42)
  rf_model.fit(x_train,y_train)
  rf_pred=rf_model.predict(x_test)
  accu=accuracy_score(y_test,rf_pred)
  print("Random Forest with",n,"number of trees having accuracy",accu)

  # Check if this model has the best accuracy so far
  if accu > best_accuracy:
      best_n = n
      best_accuracy = accu
      best_rf_model = rf_model
      best_rf_pred=rf_pred

# Print details of the best model
print(f"\nBest model details:")
print(f"Number of trees: {best_n}")
print(f"Accuracy: {best_accuracy:.4f}")


print(f"Feature importances: {best_rf_model.feature_importances_}")
```

```python
best_rf_pred
```
